```python
import cv2
from cv2 import imread
from cv2 import waitKey
from matplotlib import pyplot as plt

img = imread("img/neg1.jpg")
img_src = cv2.resize(img, (480,720))
img_src = img
plt.imshow(cv2.cvtColor(img_src,cv2.COLOR_BGR2RGB))

# print(f"hsv:{img_hsv[0][0]}")
# print(f"src:{img_src[0][0]}")
```

```python

img_hsv = cv2.cvtColor(img_src,cv2.COLOR_BGR2HSV)

score = 0.0
cnt =0
for i,row in enumerate(img_hsv):
    for j,pixel in enumerate(row):
        cnt+=1
        sc = (float(pixel[1])+float(pixel[2]))/2.0
        score+= sc
plt.imshow(cv2.cvtColor(img_src,cv2.COLOR_BGR2RGB))
plt.title(score/float(cnt))
print(score/float(cnt))
```
